```python
print('Bem-vindo a loja de Atacado da Francine')

# Solicitando ao usuário os valores
valor = float(input('Entre com o valor do produto: '))
quant = int(input('Entre com a quantidade do produto: '))
# Calculando o valor total da compra
total = valor * quant

# para valor total igual ou maior que R$ 2500 e menor que R$ 6000 desconto de 4%
# para valor total igual ou maior que R$ 6000 e menor que R$ 10000 desconto de 7%
# para valor total igual ou maior que R$ 10000 desconto de 11%
# para valor total menor que R$ 2500 não tem desconto

# Calculando o valor do desconto para cada intervalo
if (total >= 2500 and total < 6000):
    desc = total * 4/100
    final = total - desc
    print(f'O valor SEM desconto: R${total:.2f}')
    print(f'O valor COM desconto: R${final:.2f}')
elif (total >= 6000 and total < 10000):
    desc = total * 7/100
    final = total - desc
    print(f'O valor SEM desconto: R${total:.2f}')
    print(f'O valor COM desconto: R${final:.2f}')
elif (total >= 10000):
    desc = total * 11/100
    final = total - desc
    print(f'O valor SEM desconto: R${total:.2f}')
    print(f'O valor COM desconto: R${final:.2f}')
else:
    print(f'O valor SEM desconto: R${total:.2f}')
    print(f'Você não atingiu o valor mínimo para gerar desconto.')
```

```python
print('Bem-vindo a loja de Gelados da Francine')
print(20 * '-' + 'Cardápio' + 20 * '-')
print(48 * '-')
print('---| Tamanho  |  Cupuaçu (CP)  | Açaí (AC)  |---')
print('---|    P     |    R$  9.00    |  R$ 11.00  |---')
print('---|    M     |    R$ 14.00    |  R$ 16.00  |---')
print('---|    G     |    R$ 18.00    |  R$ 20.00  |---')
print(48 * '-')

valorTotal = 0.00

while True: # verificando se digitou corretamente o pedido
    sabor = input('Entre com o sabor desejado (CP/AC): ').upper()
    if (sabor != 'CP' and sabor != 'AC'):
        print('Sabor inválido. Tente novamente.\n')
        continue

    tamanho = input('Entre com o tamanho desejado (P/M/G): ').upper()
    if (tamanho != 'P' and tamanho != 'M' and tamanho != 'G'):
        print('Tamanho inválido. Tente novamente.\n')
        continue

    if (sabor == 'CP'): # realizando o pedido do cliente
        if tamanho == 'P':
            valorItem = 9.00
            print('Você pediu um Cupuaçu tamanho P: R$ 9,00\n')
        elif tamanho == 'M':
            valorItem = 14.00
            print('Você pediu um Cupuaçu tamanho M: R$ 14,00\n')
        elif tamanho == 'G':
            valorItem = 18.00
            print('Você pediu um Cupuaçu tamanho G: R$ 18,00\n')
    elif (sabor == 'AC'):
        if tamanho == 'P':
            valorItem = 11.00
            print('Você pediu um Açaí tamanho P: R$ 11,00\n')
        elif tamanho == 'M':
            valorItem = 16.00
            print('Você pediu um Açaí tamanho M: R$ 16,00\n')
        elif tamanho == 'G':
            valorItem = 20.00
            print('Você pediu um Açaí tamanho G: R$ 20,00\n')

    valorTotal += valorItem # acumulador para somar os valores do pedido

    verif = input('Deseja pedir mais alguma coisa? (S/N): ').upper()
    if (verif != 'S'): # verificando se deseja pedir algo mais
        print(f'\nO valor total a ser pago é: R$ {valorTotal:.2f}.\n')
        break
```

```python
print('Bem-vindo a Copiadora da Francine!\n')

def escolha_servico():
    while True: # Solicitando ao cliente o tipo de serviço
        print('Entre com o tipo de serviço desejado')
        print('DIG - Digitalizaçao')
        print('ICO - Impressão Colorida')
        print('IPB - Impressão Preto e Branco')
        print('FOT - Fotocópia')
        tipo = input('>> ').upper() # verificando se digitou corretamente o pedido
        if (tipo != 'DIG' and tipo != 'ICO' and tipo != 'IPB' and tipo != 'FOT' ):
            print('Escolha errada, entre com o tipo de serviço novamente.\n')
            continue
        elif (tipo == 'DIG'): # retornando o valor do serviço
            return 1.10
        elif (tipo == 'ICO'):
            return 1.00
        elif (tipo == 'IPB'):
            return 0.40
        else:
            return 0.20

def num_pagina():
    while True: # Solicitando ao cliente o número de páginas
        try:
            qtde = int(input('Entre com o número de páginas: '))
        except:
            print('O número informado deve ser inteiro, digite novamente.')
            continue

        if qtde < 20:
            return qtde
        elif qtde >= 20 and qtde < 200:
            return qtde - (qtde * 15/100)
        elif qtde >= 200 and qtde < 2000:
            return qtde - (qtde * 20/100)
        elif qtde >= 2000 and qtde < 20000:
            return qtde - (qtde * 25/100)
        else:
            print('Não aceitamos tantas páginas de uma vez. \nPor favor, entre com o número de páginas novamente.\n')

def servico_extra():
    while True: # Solicitando ao cliente se deseja algum serviço extra
        print('Deseja adicionar algum serviço?')
        print('1 - Encadernação simples - R$ 15,00')
        print('2 - Encadernação capa dura - R$ 40,00')
        print('0 - Não desejo mais nada')
        ext = (input('>> ')) # verificando se digitou corretamente
        if (ext != '1' and ext != '2' and ext != '0'):
            print('Escolha errada, entre com o tipo de serviço extra novamente.\n')
            continue
        elif (ext == '1'): # retornando o valor do serviço extra
            return 15.00
        elif (ext == '2'):
            return 40.00
        elif (ext == '0'):
            return 0.00

# realizando os cálculos para informar ao cliente o total
servico = escolha_servico()
num = num_pagina()
extra = servico_extra()

total = (servico * num) + extra

print(f'Total = R$ {total:.2f} (serviço: {servico:.0f} * páginas: {num:.0f} + extra: {extra:.2f})')
```

```python
print('-- Bem-vindo a Livraria da Francine --')
print('------------------------------------------------')

lista_livro = [] # criando uma lista de dicionários

id_global = 0  # iniciando ID do 0

# Função para cadastrar livro
def cadastrar_livro(id):
   print('------------------------------------------------')
   print('------------ MENU CADASTRAR LIVRO --------------')
   print(f'ID do livro: {id_global}')
   nome = input('Digite o nome do livro: ')
   autor = input('Digite o autor do livro: ')
   editora = input('Digite a editora do livro: ')
   livro = {
       'id': id,
       'nome': nome,
       'autor': autor,
       'editora': editora
   }

   lista_livro.append(livro.copy())  # Copiando o dicionário para dentro da lista_livro

# Função para consultar livros
def consultar_livro():
   while True:
       print('------------------------------------------------')
       print('------------ MENU CONSULTAR LIVRO --------------')
       print('Escolha a opção desejada:')
       print('1 - Consultar todos os livros')
       print('2 - Consultar livro por ID')
       print('3 - Consultar livro por autor')
       print('4 - Retornar')

       opcao = input('Escolha uma opção (1, 2, 3 ou 4): ')

       if opcao == '1':  # Consultar Todos
           print('\nLista de livros:')
           for livro in lista_livro:
               print(livro)

       elif opcao == '2':  # Consultar por Id
           id_consulta = int(input('Digite o ID do livro: '))
           encontrado = False
           for livro in lista_livro:
               if livro['id'] == id_consulta:
                   print('Livro encontrado: ')
                   print(livro)
                   encontrado = True
                   break
           if not encontrado:
               print('Livro com ID {} não encontrado.'.format(id_consulta))

       elif opcao == '3':  # Consultar por autor
           autor_consulta = input('Digite o nome do autor para consultar: ')
           encontrado = False
           for livro in lista_livro:
               if livro['autor'].lower() == autor_consulta.lower():
                   if not encontrado:
                       print('Livro do autor {}: '.format(autor_consulta))
                   print(livro)
                   encontrado = True
           if not encontrado:
               print('Nenhum livro do autor encontrado {}.'.format(autor_consulta))

       elif opcao == '4':  # Retornar ao menu principal
           return

       else:
           print("Opção inválida. Escolha uma opção válida.")

# Função para remover o livro
def remover_livro():
   while True:
       id_remover = int(input("Digite o id do livro a ser removido: "))
       encontrado = False
       for livro in lista_livro:
           if livro["id"] == id_remover:
               lista_livro.remove(livro)
               print("Livro removido com sucesso.")
               encontrado = True
               break
       if not encontrado:
           print("Id {} inválido. Livro não encontrado.".format(id_remover))
       else:
           break

# Criando o Menu principal
while True:
   print('------------------------------------------------')
   print('---------------- MENU PRINCIPAL ----------------')
   print('Escolha a opção desejada:')
   print("1 - Cadastrar Livro")
   print("2 - Consultar Livro")
   print("3 - Remover Livro")
   print("4 - Sair")

   opcao_principal = input("Escolha uma opção (1, 2, 3 ou 4): >> ")

   if opcao_principal == '1':  # Cadastrar livro
       id_global += 1
       cadastrar_livro(id_global)

   elif opcao_principal == '2':  # Consultar livro
       consultar_livro()

   elif opcao_principal == '3':  # Remover livro
       remover_livro()

   elif opcao_principal == '4':  # Encerrar Programa
       print("Encerrando o programa...")
       break

   else:
       print("Opção inválida. Escolha uma opção válida.")
```
